```python
import pandas as pd
```

```python
#Load in the data
df = pd.read_csv("data/Lobbyists__Fundraising_and_Political_Consulting_Reports_20250613.csv")
df.head()
```

```python
# Analyze the data for missing info and statistical information
df.isnull().sum()
```

```python
#Drop the third party since no information exists in such category and the additional information about the firm itself
df.columns
df.drop_duplicates()
lobbying = df.copy().drop(["Filing Period", "Street Address", "City", "State", "Zip Code", "Phone Number","Principal First Name", 
                "Principal Last Name", "Lobbyist Employees", "Political Committee", "Third Party"], axis = 1)
lobbying.head(5)
```

```python
#Impute data regarding the target office
lobbying["Target Office"] = lobbying["Target Office"].replace("- Select -", "Mayor")
```

## Which government offices have attracted the most fundraising?

```python
lobbying.sort_values("Total Raised", ascending = False)["Target Office"].head(10)

# Findings: The office of the Mayor and the Attorney General were associated with the largest
# fundraising amounts, indicating a trend of higher political contributions towards these offices. 
# This trend is reflected in the high-profile of these offices, attracting more attention and 
# ultimately higher fundraising efforts. 
```

 ## Which fundraising groups were most active? Who were their clients?

```python
fundraising = lobbying[lobbying["Report Type"] == "Fundraising"].sort_values("Total Raised", ascending = False)
fundraising.head(5)

#Findings: Greenberg Traurig, LLP and Verizon Corporate Resources Group LLC were the biggest fundraisers. 
```

```python
#Companies whose compensation Owed == Total Raised (can be attributed to how they report, all the funds raised
#are compensation to the firms, or was volunteer work)
lobbying[(lobbying["Compensation Owed"] == lobbying["Total Raised"])
& (lobbying["Report Type"] == "Fundraising")].groupby("Entity Name").size().sort_values(ascending = False)
```

## Are some offices (e.g. Mayor) more likely to use fundraising groups than others?

```python
lobbying[lobbying["Report Type"] == "Fundraising"]["Target Office"].value_counts()

## Findings: The City Council office uses more fundraising groups, but this could be attributed to the pure quantity
# of them over other offices, such as the Mayor. 
```
